```python
#Import các thư viện cần thiết
import matplotlib
import numpy as np
import numpy.matlib
import random
import PIL
import os
# import warnings
# warnings.filterwarnings("ignore")
```

```python
# Đọc hiển thị bức ảnh đầu vào
image = PIL.Image.open(path_image)
print("Ảnh đầu vào")
matplotlib.pyplot.imshow(image)
```

```python
image = np.array(image)
print(image.shape)
```

```python
#1. Thay đổi độ sáng cho ảnh (1 điểm)
#Điều chỉnh độ sáng có nghĩa là tăng giá trị pixel đồng đều trên tất cả các kênh cho toàn bộ hình ảnh để tăng độ sáng hoặc giảm giá trị pixel đồng đều trên tất cả các kênh để toàn bộ hình ảnh giảm độ sáng.
#Tăng độ sáng làm cho hình ảnh trắng hơn. Và giảm độ sáng làm tối hình ảnh.
beta = 50 # giá trị beta thể hiện độ sáng của ảnh, beta càng lớn thì ảnh càng sáng.
new_image = np.zeros(image.shape, image.dtype)
for y in range(image.shape[0]):
  for x in range(image.shape[1]):
     for c in range(image.shape[2]):
          new_image[y,x,c] = np.clip(image[y,x,c] + beta, 0, 255)
matplotlib.pyplot.imshow(new_image)
```

```python
# 2. Điều chỉnh độ tương phản của ảnh
alpha = 1.5 #giá trị alpha để điều chỉnh độ tương phản, nếu giá trị alpha = 1.0 là ảnh bình thường
new_image = np.zeros(image.shape, image.dtype)
for y in range(image.shape[0]):
  for x in range(image.shape[1]):
     for c in range(image.shape[2]):
          new_image[y,x,c] = np.clip(alpha*image[y,x,c], 0, 255)
matplotlib.pyplot.imshow(new_image)
```

```python
#3. Chuyển đổi ảnh màu sang ảnh xám
def convert_image2_grayscale(image):
    rgb_weights = [0.2989, 0.5870, 0.1140]
    image2 = np.array(image).astype(float)
    grayscale_image = rgb_weights[0] * image2[:,:,0] + rgb_weights[1] * image2[:,:,1] + rgb_weights[2] * image2[:,:,2]
    grayscale_image = np.dstack([grayscale_image,grayscale_image,grayscale_image]).astype(np.uint8)
    return grayscale_image
grayscale_image = convert_image2_grayscale(image)
matplotlib.pyplot.imshow(grayscale_image)
```

```python
#6. Làm mở ảnh sử dụng GAUSIAN FILTER
# gaussian filter
import math
# Công thức của gaussian . Đường dẫn tham khảo: https://en.wikipedia.org/wiki/Gaussian_blur
def gaussianFormular(x,y,sigma):
    return (math.exp(-(x**2 + y**2)/(2*(sigma**2))))/(2*3.14*(sigma**2))
# 
def gaussFilter(size, sigma):
    output = np.zeros(size)
    for i in range(size[0]):
        for j in range(size[1]):
            output[i,j] = gaussianFormular(i-size[0]//2,j-size[1]//2, sigma)
    return output/np.sum(output)

def convolution(image, filter):
    iw,ih,id = image.shape
    fw,fh = filter.shape 
    out = np.zeros((iw-fw+1,ih-fh+1,id))
    for d in range(id):
        for w in range(ih-fh+1):
            for h in range(iw-fw+1):
                out[w,h,d] = np.sum(filter*image[w:w+fh , h:h+fw , d])
    
    if id == 1:
        return np.resize(out, (out.shape[0], out.shape[1])).astype(np.uint8)
    else:
        return out.astype(np.uint8)
# Khởi tạo hàm Gaussian Filder kích thước 3x3
gaussianFilter = gaussFilter((3,3),5)
# Chạy bộ lọc convolution trên ảnh. Tham khảo:https://en.wikipedia.org/wiki/Kernel_(image_processing)
gaussian_image = convolution(image, gaussianFilter)
#Hiển thị ảnh áp dụng gassianFilter
matplotlib.pyplot.imshow(gaussian_image)
```

```python
# 5. Chồng 2 ảnh cùng kích thước (2 điểm): chỉ làm trên ảnh xám
import math
from PIL import Image

image1_path = 'dog.jpg'
image1 = Image.open(image1_path)
print("Show hình image 1")
matplotlib.pyplot.imshow(image1)
```

```python
image2_path = 'meo.jpg'
image2 = Image.open(image2_path)
print("Show hình image2")
matplotlib.pyplot.imshow(image2)
```

```python
# Chồng 2 hình với nhau
# Đầu tiên chuyển ảnh  về dạng ảnh xám (grayscale image)
# If you have an L mode image, that means it is a single channel image - normally interpreted as greyscale. The L means that is just stores the Luminance. It is very compact, but only stores a greyscale, not colour 
#https://stackoverflow.com/questions/52307290/what-is-the-difference-between-images-in-p-and-l-mode-in-pil

# Sau đó resize 2 ảnh cùng kích thước với nhau
def blend_image(image1, image2,theshold=0.5):
  image2 = image2.resize(image1.size)
  image1 = convert_image2_grayscale(image1)
  image2 = convert_image2_grayscale(image2)

  result = (theshold*image1 + theshold*image2).astype(np.uint8)
  return result

print("Show kết quả chèn hai ảnh với nhau")
result = blend_image(image1, image2, 0.5)
matplotlib.pyplot.imshow(result)
```
